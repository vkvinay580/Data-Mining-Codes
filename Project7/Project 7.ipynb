```python
import random
```

```python
Attacker_dices = 3
Defender_dices = 2
outcome_dict = {'attacker':0, 'both':0, 'defender':0}
number_of_turns = int(input("Please insert number of turns: 10, 100, 10000 ...   "))
```

```python
def risk_game(Attacker_dices,Defender_dices ):
    attacker_scores = []
    for i in range(Attacker_dices):
        score = random.randint(1,6)
        attacker_scores.append(score)

    defender_scores = []
    for i in range(Defender_dices):
        score = random.randint(1,6)
        defender_scores.append(score)

  
    attacker_highest_score = max(attacker_scores)
    defender_highest_score = max(defender_scores)


    if attacker_highest_score > defender_highest_score:
        Defender_dices -= 1
    else:
        Attacker_dices -= 1
         
    return  Attacker_dices, Defender_dices
    
```
